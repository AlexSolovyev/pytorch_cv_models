```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

from PIL import Image

import os

import torch
from torch import nn
from torch.nn import functional as F
```

```python
class VGG11(nn.Module):

    def __init__(self):
        """Initialize VGG11 params."""

        super().__init__()
        
        nn.Sequential(
            nn.Conv2d(3, 64, 3),     # (3, 224, 224) -> (64, 222, 222)
            nn.MaxPool2d(),          # (64, 222, 222) -> (64, 111, 111)
            nn.Conv2d(64, 128, 3),   # (64, 111, 111) -> (128, 109, 109)
            nn.MaxPool2d(),          # (128, 109, 109) -> (128, 55, 55)
            nn.Conv2d(128, 256, 3),  # (128, 55, 55) -> (256, 53, 53)
            nn.Conv2d(256, 256, 3),  # (256, 53, 53) -> (256, 51, 51)
            nn.MaxPool2d(),          # (256, 51, 51) -> (256, 26, 26)
            nn.Conv2d(256, 512, 3),  # (256, 26, 26) -> (512, 24, 24)
            nn.Conv2d(512, 512, 3),  # (512, 24, 24) -> (512, 22, 22)
            nn.MaxPool2d(),          # (512, 22, 22) -> (512, 11, 11)
            nn.Conv2d(512, 512, 3),  # (512, 11, 11) -> (512, 9, 9)
            nn.Conv2d(512, 512, 3),  # (512, 9, 9) -> (512, 512, 7)
            nn.MaxPool2d(),          # (512, 9, 9) -> (512, 512, 5)
            nn.Linear(4096),         # * -> 4096
            nn.Linear(4096),         # 4096 -> 4096
            nn.Linear(1000),         # 4096 -> 1000
        )

    def forward(self, x):
        """Forward pass for VGG11."""
        return self.VGG11(x)
```
